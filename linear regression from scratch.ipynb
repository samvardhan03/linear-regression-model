```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('brainhead.csv')
```

```python
df.head()
```

```python
X=df['headsize'].values
Y = df['brainweight'].values
```

```python
np.corrcoef(X, Y)

plt.scatter(X, Y, c='blue', label='Data points')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
#values
n = len(X)
```

```python
# calculate theta1 and theta2
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
print("coefficients for regression",b1, b0)
```

```python
#%matplotlib inline
plt.show()
plt.rcParams['figure.figsize'] = (10.0, 5.0)

y = b0 + b1 * X
```

```python
# Ploting Line
plt.plot(X, y, color='blue', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='red', label='Scatter data')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```
